## B. Processing function


```python
from PIL import Image
import numpy as np

def process_image(fpath):

    image = Image.open(fpath).convert('RGB')
    image_array = np.ascontiguousarray(image)
    
    #image_show(image)
    width, height, ch = image_array.shape
    if width < 300 or height < 300:
        return image, False  # false for image invalid
    
    if ch == 4: # or #if image.mode == "RGBA": image .convert("RGB")
        image = Image.fromarray(image_array).convert('RGB')
    
    return image, True # true for image ok
```

```python
import os
from genericpath import exists
import glob
import cv2

dest_folder = os.path.join(root_dir,'processed_images')
if not os.path.exists(dest_folder):
    os.mkdir(dest_folder)

inval_folder = os.path.join(root_dir,'invalid_images')
if not os.path.exists(inval_folder):
    os.mkdir(inval_folder)
```

## C. Looping over the files and processin

```python
sub_dirs = glob.iglob(f'{base_dir}/*')
tot_count_val = 0
tot_count_inv = 0
for subdir in sub_dirs:
    if os.path.isdir(subdir):
        print(f'********PROCESSING {subdir.split("/")[-1]} *********')
        class_folders = glob.iglob(f'{subdir}/*') 
    
    cls_count_val=0
    cls_count_inv=0
        
    for cls in class_folders:
        if os.path.isdir(cls):
            print(f'>> {cls.split("/")[-1]} ')
            files = glob.iglob(f'{cls}/*.*')
        
        class_name = (cls.split('/')[-1]).lower().strip()
        class_name_new = class_name
        if class_name[-1]=='s':    # corrections for class_name plurism
            class_name_new = class_name[:-1]
        
        count = 0
        for f in files:
            if os.path.isfile(f):
                image, flag = process_image(f)  # process image
                file_name = f.split('/')[-1]
                f_name, f_ext = os.path.splitext(file_name)
                file_name_jpg = f'{f_name}.jpg'  # for later conversion to jpg
                
                print(flag,f)
                if (flag):
                    tot_count_val += 1; cls_count_val += 1; count += 1
                    class_dir = os.path.join(dest_folder, class_name_new)
                    if not os.path.exists(class_dir):
                        os.mkdir(class_dir)

                    dest_file = os.path.join(dest_folder, class_name_new, file_name_jpg)
                    image.save(dest_file)
                    
                else:
                    tot_count_inv += 1; cls_count_inv += 1;
                    class_dir = os.path.join(inval_folder, class_name_new)
                    if not os.path.exists(class_dir):
                        os.mkdir(class_dir)

                    inval_file = os.path.join(inval_folder, class_name_new, file_name_jpg)
                    image.save(inval_file)
            
        print(count)    
    print(f'Total files saved.. valid={cls_count_val}, invalid={cls_count_inv}')
    print('\n')
print('\nXXXXXXXXXXXXXXXXXXXXXXXXX   FINISHED   XXXXXXXXXXXXXXXXXXXXXXXXX')
print(f'Total files saved.. valid={tot_count_val}, invalid={tot_count_inv}')
```
