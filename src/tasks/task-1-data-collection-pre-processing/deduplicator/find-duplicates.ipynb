```python
import matplotlib as mpl
from matplotlib import pyplot as plt
from matplotlib.image import imread

from main import run

mpl.rcParams['figure.dpi'] = 200
```

```python
def display_images_horizontally(ref_img, img_list):
    num_of_images = len(img_list)
    
    fig = plt.figure()
    
    # display reference image
    a = fig.add_subplot(1, num_of_images, 1)
    plt.gca().set_title('ref_img')
    plt.imshow(plt.imread(ref_img), cmap='Greys_r')
    plt.axis('off')
    
    # display duplicates
    for i in range(1, num_of_images):
        a = fig.add_subplot(1, num_of_images, i+1)
        plt.imshow(plt.imread(img_list[i]), cmap='Greys_r')
        plt.axis('off')
```

```python
duplicates = run("/Users/ugur/Documents/projects/pothole-detection/output/frames/*/*")
```

```python
DISPLAY_MAX_N_DUPLICATES = 5

for ref_img_path, duplicate_image_paths in duplicates.items():
    display_images_horizontally(ref_img_path, duplicate_image_paths[:DISPLAY_MAX_N_DUPLICATES])
```
